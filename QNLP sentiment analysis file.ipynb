```python
from bs4 import BeautifulSoup as bs
import requests #to access website
import pandas as pd
```

```python
r = requests.get("https://www.imdb.com/title/tt0111161/reviews/?ref_=tt_ov_rt")

# Convert to a beautiful soup object
soup = bs(r.content)

# Print out HTML
contents = soup.prettify()
print(contents[:150])
```

```python
movie_title = []
movie_review = []
movie_rating = []
```

```python
reviews=soup.find_all("div",class_="lister-item-content")
for ele in reviews:
  review_text = ele.find("a", class_="title").text.strip()
  movie_review.append(review_text)
movie_review
```

```python
imdb_subpge_title = soup.find(class_="subpage_title_block__right-column")
imdb_subpge_title_1=imdb_subpge_title.find_all(class_="parent")
```

```python
for row in imdb_subpge_title_1:
  title=row.a.text
  movie_title.append(title)
movie_title
```

```python
from textblob import TextBlob
textfile=open("Sentanalysis.txt",'r')
sent=[]
for line in textfile:
  blob=TextBlob(line)
  sentiment=blob.sentiment.polarity
  sent.append(sentiment)
sent
```

```python
sent2=[]
for i in sent:
  if i>0:
    sent2.append(1)#positive
  else:
    sent2.append(0)#negative
sent2
```

```python
finalfile=open("finaltext.txt",'w')
for i in range(0,249):
  finalfile.write(str(i))
  finalfile.write(' , ')
  finalfile.write(str(sent2[i]))
  finalfile.write(' , ')
  finalfile.write(movie_review[i]+'\n')
finalfile=open("finaltext.txt",'r')
print(finalfile.read())
```
